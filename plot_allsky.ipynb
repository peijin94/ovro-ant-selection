```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
fname_set = './*image.fits'
fnames = glob(fname_set)

for fname in fnames:
    hdul = fits.open(fname)
    data = hdul[0].data
    print(data.shape)
    hdul.close()
    plt.figure()
    plt.imshow((data.squeeze()), cmap='gray',vmax=0.3*np.nanmax(data))
    plt.title(fname)
```
